```python
for y in range(2000,2101):
    try:
        d = date(y,2,29)
        print(y, "is a leap year")
    except ValueError as e:
        print(y, "is not a leap year")
```

```python
print(datetime(2000, 2, 29))
print(datetime(2000, 2, 29, 12))
print(datetime(2000, 2, 29, 12, 30))
print(datetime(2000, 2, 29, 12, 30, 24))
print(datetime(2000, 2, 29, second=24))
```

```python
dateStrFr='29/02/2020'
dateStrEn='02/29/2020'
dateStrIso='2020-02-29'
```

```python
d = date.fromisoformat(dateStrIso)
d
```

```python
# https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
d2 = datetime.strptime(dateStrFr, "%d/%m/%Y").date()
d3 = datetime.strptime(dateStrEn, "%m/%d/%Y").date()
print(d2,d3)
```

```python
print(dt1)
print(dt1.strftime('%H:%M:%S %d/%m/%Y'))
print(dt1.strftime('%A %d %B %Y %H:%M'))
```

```python
locale.setlocale(locale.LC_ALL, 'fr_FR')
print(dt1.strftime('%A %d %B %Y %H:%M'))
locale.setlocale(locale.LC_ALL, 'en_UK')
print(dt1.strftime('%A %d %B %Y %H:%M'))
locale.setlocale(locale.LC_ALL, 'es_ES')
print(dt1.strftime('%A %d %B %Y %H:%M'))
```

```python
# with babel
for l in babel.localedata.locale_identifiers():
    try:
        locale.setlocale(locale.LC_ALL, l)
        print(l, "->", dt1.strftime('%A %d %B %Y %H:%M'))
    except:
        print(l, "not available")
```

```python
print(datetime.now(tz=pytz.timezone('Pacific/Kiritimati')))
print(datetime.now(tz=pytz.timezone('Pacific/Honolulu')))
print(datetime.now(tz=pytz.timezone('America/Los_angeles')))
print(datetime.now(tz=pytz.timezone('America/New_york')))
print(datetime.now(tz=pytz.timezone('Europe/Paris')))
```
