## Données météo Toulouse

```python
# download data here: https://data.toulouse-metropole.fr/explore/dataset/51-station-meteo-toulouse-lardenne/information/

dfMeteo = pd.read_csv('datas/51-station-meteo-toulouse-lardenne.csv', sep=';', 
                      parse_dates=['heure_de_paris','heure_utc'])
dfMeteo.head(5)
```

```python
dfMeteo.info()
```

```python
dfMeteo.loc[0, 'heure_de_paris']
```

```python
# operator elementwise for pandas & numpy: & (and), | (or), ~ (not), ^ (xor)
dfMeteoJuly2022 = dfMeteo[(dfMeteo.heure_utc.dt.year == 2022) & (dfMeteo.heure_utc.dt.month == 7)]
dfMeteoJuly2022.head(3)
```

```python
dfMeteo.temperature.isna().any()
```

```python
dfMeteo.loc[0, 'temperature'] = np.nan
```

```python
dfMeteo.temperature.isna().any()
```

```python
dfMeteo[dfMeteo.temperature.isna()]
```

```python
dfMeteo[~dfMeteo.temperature.isna()]
```

```python
dfMeteoIndexTime =  dfMeteo.set_index('heure_utc').sort_index()
dfMeteoIndexTime
```

```python
dfMeteoIndexTime.loc['2022-07']
```

```python
dfMeteoSummer2022 = dfMeteoIndexTime.loc['2022-07':'2022-08']
dfMeteoSummer2022
```

```python
dfMeteoSummer2022.temperature.plot()
```

```python
g = dfMeteoSummer2022.groupby(pd.Grouper(freq='1D'))
```

```python
maxTempJuly = g.temperature.max()
maxTempJuly
```

```python
plt.plot(maxTempJuly)
```

```python
plt.plot(g[['temperature', 'pluie_intensite_max']].max())
# g[['temperature', 'pluie_intensite_max']].max().plot()
```

```python
stats = g[['temperature','pression', 'pluie_intensite_max']].agg(['min','max','mean','median'])
stats
```

```python
plt.plot(stats['temperature'])
```

```python
stats2 = g.agg({'temperature':['min','max','mean','median'], 'pluie_intensite_max': ['max', np.ptp]})
stats2
```
